##Creación de variables

```python
def clasificar_ram(ram):
    if ram < 3:
        return "Bajo"
    elif ram == 4:
        return "Medio-bajo"
    elif ram==6:
        return "Medio"
    elif ram==8:
        return "Medio-alto"
    elif ram==12:
        return "Alto"
    elif ram>12:
        return "Premium"


df['Calidad según Ram'] = df['ram_gb'].map(clasificar_ram)
df.head()

```

```python
df['Eficiencia bateria'] = round(df['battery_mah']/df["charging_watt"])
df.head()
```

```python
df['Porcentaje eficiencia']= round((df['Eficiencia bateria']*100)/333)
df.head()
```

#Filtrado y condiciones en Pandas

```python
#Celulares con calidad de ram Premium o Alta y tienen una calificación mayor o igual a 4.
df[((df['Calidad según Ram']== 'Premium') | (df['Calidad según Ram']=="Alto")) & (df["rating"]>=4)]
```

```python
#Celulares con un valor mayor a 815, que soporte 5G y tengan capacidad almacenamiento mayor a 128GB
df[(df['price_usd']>815)&(df['5g_support']=="Yes")&(df['storage_gb']>128)]
```

```python
#Celulares marca Samsung o Apple y una calificación mayor a 4.5.
x=df[((df['brand']=="Samsung")|(df['brand']=="Apple"))&(df['rating']>4.5)]
x
```

```python
#Celulares que fueron lanzados al mercado en Noviembre y que la calidad de ram sea Premium.
df[(df['release_month']=="November")&(df['Calidad según Ram']=="Premium")]
```

```python
#Las 3 marcas con los mejores procesadores y con un porcentaje de eficiencia mayor a 75.
df[((df['processor']=="Snapdragon 8 Gen 3")| (df['processor']=="A18 Pro")| (df['processor']=="Dimensity 9300"))& (df['Porcentaje eficiencia']>75) ]
```

#Visualizaciones:

```python
#Diagrama de barras del promedio de los precios de las marcas.
mean_prices = df.groupby("brand")["price_usd"].mean()

plt.figure(figsize=(8,5))
plt.bar(mean_prices.index, mean_prices.values)
plt.xlabel("Brand")
plt.ylabel("Precio Promedio (USD)")
plt.title("Precio promedio por marca")
plt.tight_layout()
plt.show()
```

```python
#Mapa de calor del precio de los celulares teniendo en cuenta su marca y mes de lanzamiento.
grouped = df.groupby(["brand", "release_month"])["price_usd"].mean().reset_index()
pivot = grouped.pivot(index="brand", columns="release_month", values="price_usd")

plt.figure(figsize=(11,5))
plt.imshow(pivot, aspect="auto")
plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.title("Precio promedio por marca y mes de lanzamiento")
plt.colorbar(label="Precio promedio (USD)")
plt.tight_layout()
plt.show()
```

```python
#Scatter plot del precio en base a la calidad de la ram.
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(14, 8))
colors = {
    'Premium': '#FFD700',
    'Alto': '#FF6B6B',
    'Medio-alto': '#4ECDC4',
    'Medio': '#95E1D3',
    'Medio-bajo': '#A8E6CF',
    "Bajo":"#0525EEB"
}

for segment in df['Calidad según Ram'].unique():
    segment_data = df[df['Calidad según Ram'] == segment]
    ax.scatter(segment_data['ram_gb'],
               segment_data['price_usd'],
               s=80,
               alpha=0.5,
               c=colors[segment],
               label=segment,
               edgecolors='none')

ax.set_xlabel('RAM (GB)', fontsize=13, fontweight='bold')
ax.set_ylabel('Precio (USD)', fontsize=13, fontweight='bold')
ax.set_title('Relación entre Precio y RAM según Segmento de Calidad\n(n=1000 smartphones)',
             fontsize=15, fontweight='bold', pad=20)
ax.legend(title='Segmento', title_fontsize=12, fontsize=11,
          loc='upper left', framealpha=0.9)
ax.grid(True, alpha=0.3)

z = np.polyfit(df['ram_gb'], df['price_usd'], 1)
p = np.poly1d(z)
ax.plot(df['ram_gb'], p(df['ram_gb']),
        "r--", alpha=0.7, linewidth=3, label='Tendencia general')

plt.tight_layout()
plt.show()
```

```python
#Diagrama de barras del costo promedio en base al almacenamiento(GB) por marca.
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(14, 7))

all_brands = df['brand'].unique()

colors_list = plt.cm.tab20(np.linspace(0, 1, len(all_brands)))
brand_colors = {brand: colors_list[i] for i, brand in enumerate(all_brands)}

known_colors = {
    'Apple': '#A2AAAD',
    'Samsung': '#1428A0',
    'Google': '#4285F4',
    'OnePlus': '#F50514',
    'Oppo': '#00A862',
    'Infinix': '#FF6B35',
    'Realme': '#FFCC00',
    'Xiaomi': '#FF6900',
    'Motorola': '#0066CC',
    'Vivo': '#1157B0'
}

brand_colors.update({k: v for k, v in known_colors.items() if k in all_brands})

df['precio_por_gb'] = df['price_usd']

top_brands = df['brand'].value_counts().head(15).index
df_top = df[df['brand'].isin(top_brands)]

precio_por_marca = df_top.groupby('brand')['precio_por_gb'].mean().sort_values(ascending=False)

bars = ax.bar(range(len(precio_por_marca)),
              precio_por_marca.values,
              color=[brand_colors.get(brand, '#808080') for brand in precio_por_marca.index],
              alpha=0.8,
              edgecolor='black',
              linewidth=1.5)

for i, (brand, value) in enumerate(precio_por_marca.items()):
    ax.text(i, value + (precio_por_marca.max() * 0.02),
            f'${value:.2f}',
            ha='center', va='bottom', fontweight='bold', fontsize=10)

ax.set_xlabel('Marca', fontsize=13, fontweight='bold')
ax.set_ylabel('Precio en USD', fontsize=13, fontweight='bold')
ax.set_title('Costo Promedio por GB de Almacenamiento por Marca\n(Top 15 marcas - n=1000 smartphones)',
             fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(range(len(precio_por_marca)))
ax.set_xticklabels(precio_por_marca.index, rotation=45, ha='right', fontsize=10)
ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
#Calificación promedio en base al procesador.
sns.set_style("whitegrid")
fig, ax = plt.subplots(figsize=(12, 10))

top_processors = df['processor'].value_counts().head(15).index
df_top = df[df['processor'].isin(top_processors)]

rating_por_procesador = df_top.groupby('processor')['rating'].mean().sort_values(ascending=False)

colors = plt.cm.RdYlGn(np.linspace(0.3, 0.9, len(rating_por_procesador)))

bars = ax.barh(range(len(rating_por_procesador)),
               rating_por_procesador.values,
               color=colors,
               alpha=0.8,
               edgecolor='black',
               linewidth=1.5)

for i, (processor, value) in enumerate(rating_por_procesador.items()):
    count = df_top[df_top['processor'] == processor].shape[0]
    ax.text(value + 0.05, i,
            f'{value:.2f} ⭐ (n={count})',
            va='center', fontweight='bold', fontsize=10)

ax.set_ylabel('Procesador', fontsize=13, fontweight='bold')
ax.set_xlabel('Rating Promedio', fontsize=13, fontweight='bold')
ax.set_title('Rating Promedio por Procesador\n(Top 15 procesadores más comunes - n=1000 smartphones)',
             fontsize=14, fontweight='bold', pad=20)
ax.set_yticks(range(len(rating_por_procesador)))
ax.set_yticklabels(rating_por_procesador.index, fontsize=10)
ax.set_xlim(0, 5.5)
ax.grid(True, alpha=0.3, axis='x')

ax.axvline(x=4.0, color='red', linestyle='--', linewidth=2, alpha=0.6, label='Rating 4.0 (Bueno)')
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()
```
